```python
df_cut=pd.crosstab(index=df['color'],columns=df['cut'])
df_cut
```

```python
b=df_cut.plot(kind='bar',stacked=True,xlabel='color',ylabel='cut',figsize=(4,5))
```

```python
print(b.containers[0][6])
```

```python
for c in b.containers:
    for v in c:

        labels=[v.get_height() if v.get_height()>0 else '' for v in c]
    b.bar_label(c,labels=labels)
    
```
