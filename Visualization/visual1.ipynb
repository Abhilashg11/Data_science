```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x=[1,3,5,6,7,8]
y=[100,200,230,40,129,90]
```

```python
plt.figure(figsize=(10,5),facecolor='orange')

plt.scatter(x,y,color='r',marker='+')
plt.xlabel('x',color='b',size=10)
plt.ylabel('y',color='b',size=10)
plt.title('Data science',color = 'pink',size = 20)
plt.plot(x,y,color = 'g',linewidth=2,linestyle=':')
plt.suptitle('graph',y=1,color='green',size=13)
plt.xticks(np.arange(1,20,2));
```
